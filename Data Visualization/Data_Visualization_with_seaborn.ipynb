###Inicio

```
import pandas as pd
import numpy as np
from numpy import median

import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

###Bar plot


```
tips = sns.load_dataset('tips')
```

```
tips.head(3)
```

```
sns.barplot(x='day', y='tip', data=tips, hue='sex', palette='winter_r')
```

```
sns.barplot(x='day', y='total_bill', data=tips, hue='smoker', palette='winter_r')
```

```
sns.barplot(x='total_bill', y='day', data=tips, palette='spring')
```

```
sns.barplot(x='day', y='tip', data=tips, palette='spring', order=['Sat', 'Fri', 'Sun', 'Thur'])
```

```
sns.barplot(x='day', y='total_bill', data=tips, estimator=median, palette='spring')
```

```
sns.barplot(x='smoker', y='tip', data=tips, estimator=median, hue='sex', palette='coolwarm')
```

```
sns.barplot(x='smoker', y='tip', data=tips, ci=34, palette='winter_r', estimator=median) # ci = intervalo de confianca
```

```
sns.barplot(x='day', y='total_bill', data=tips, palette='spring', capsize=0.9)
```

```
sns.barplot(x='day', y='total_bill', data=tips, hue='sex', palette='spring', capsize=0.1)
```

```
sns.barplot(x='size', y='tip', data=tips, capsize=0.15, palette='husl')
```

```
sns.barplot(x='size', y='tip', data=tips, capsize=0.15, color='red', saturation=0.8)
```

###Dist Plot



```
num = np.random.randn(150)
```

```
sns.distplot(num)
```

```
sns.distplot(num, color='red')
```

```
label_dist = pd.Series(num, name='variable x')
```

```
sns.distplot(label_dist)
```

```
sns.distplot(label_dist, vertical=True, color='red')
```

```
sns.distplot(label_dist, hist=False, rug=True, color='green')
```

```
sns.distplot(label_dist, hist=False, rug=True)
```

### Strip plots

```
sns.stripplot(x='tip' ,data=tips, color='green')
```

```
sns.stripplot(x='total_bill', data=tips)
```

```
sns.stripplot(x='day', y='total_bill', data=tips)
```

```
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True)
```

```
sns.stripplot(x='day', y='total_bill', data=tips, jitter=0.02)
```

```
sns.stripplot(x='total_bill', y='day', data=tips, jitter=1, linewidth=1.5)
```

```
sns.stripplot(x='day', y='total_bill', data=tips, hue='sex', jitter=True)
```

```
sns.stripplot(x='day', y='total_bill', data=tips, hue='sex', 
              jitter=True, split=True, palette='winter_r')
```

```
sns.stripplot(x='day', y='total_bill', data=tips, hue='smoker', 
              jitter=True, split=True, palette='winter_r', order=['Fri', 'Sun', 'Sat', 'Thur'])
```

```
sns.stripplot(x='sex', y='tip', data=tips, order=['Female', 'Male'], marker='D', palette='husl')
```

```
sns.stripplot(x='day', y='total_bill', data=tips, marker='D', size=10, hue='sex')
```

```
sns.stripplot(x='day', y='total_bill', data=tips, marker='D', size=10, hue='sex', 
              edgecolor='gray', alpha=0.8, jitter=True, split=True, palette='husl')
```

```
sns.stripplot(x='tip', y='day', data=tips)
sns.boxplot(x='tip', y='day', data=tips)
```

```
sns.stripplot(x='tip', y='day', data=tips, jitter=True, palette='winter_r', color='1s')
sns.boxplot(x='tip', y='day', data=tips)
```

```
sns.stripplot(x='tip', y='day', data=tips, jitter=True)
sns.violinplot(x='tip', y='day', data=tips, color='0.9')
```
